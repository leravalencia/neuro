```python
# import the necessary packages
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import BernoulliRBM
from sklearn.pipeline import Pipeline
import numpy as np
import argparse
import time
import pandas as pd
import pandas as pd
from sklearn import linear_model
```

```python
df_test = pd.read_csv('sign_mnist_test.csv')
```

```python
df_train = pd.read_csv('sign_mnist_train.csv')
```

```python
df_test.info()
```

```python
X_test = df_test.loc[:, 'pixel1':'pixel784']
Y_test = df_test['label']
X_train = df_train.loc[:, 'pixel1':'pixel784']
Y_train = df_train['label']
```

```python
logistic = linear_model.LogisticRegression(C=10)
rbm = BernoulliRBM(n_components=180, learning_rate=0.01, batch_size=10, n_iter=50, verbose=True, random_state=None)
clf = Pipeline(steps=[('rbm', rbm), ('clf', logistic)])
```

```python
clf.fit(X_test,Y_test) 
```

```python
Y_pred = clf.predict(X_train)
print ( classification_report(Y_train, Y_pred))
```
